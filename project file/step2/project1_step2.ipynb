```python
import numpy as np
import pandas as pd

df = df = pd.read_csv('./chocolate_preprocessed.csv')
```

```python
a=(100/df['Rating'].max())
df['Rating'] = df['Rating']*a

df.head()
```

```python
df['price(100g)'] = (df['Rating']*df['Cocoa Percent'])*25
df
```

```python
m=pd.DataFrame(df[df['Cocoa Percent']>70.0])
dark_chocolates = m.reset_index()
dark_chocolates.drop(['index'],axis=1,inplace=True)
```

```python
has_Trinitario = np.array(dark_chocolates['Bean Type']=='Trinitario')
```

```python
dark_chocolates.loc[has_Trinitario, 'price(100g)']=dark_chocolates.loc[has_Trinitario, 'price(100g)']*1.1
dark_chocolates
```

```python
priceSum = dark_chocolates['price(100g)'].sum()
```
